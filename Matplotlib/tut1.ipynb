```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# line curve

x1 = np.linspace(0,10,100)
print(x1)

fig1 = plt.figure()

plt.plot(x1,np.sin(x1),'-', label="sin curve")
plt.plot(x1,np.cos(x1),'--',label="cos curve")
#to show the label/legend
plt.legend()
plt.show()
```

```python
#line curve between two varaibles
x2 = np.arange(0,10)
y2 = np.arange(0,10)
y2 = np.square(y2)
print(x2)
print(y2)
plt.plot(x2,y2)
plt.show()
```

```python
#comparison line curve
x3 = np.arange(0,10)

plt.plot(x3,np.square(x3),label="square")
plt.plot(x3,(x3)**3,label="cube")
plt.plot(x3,(x3)**4,label="quad")
plt.legend()
plt.show()
```

```python
plt.plot(x3,np.log10(x3),label="log base 10")
plt.plot(x3,np.log2(x3),label="log base 2")
plt.legend()
plt.show()
```

```python
#scatter plot
x4 = np.linspace(10,50,10)
y4 = np.array([1,4,9,16,25,36,49,64,81,100])
plt.scatter(x4,y4)
plt.show()
```

```python
#histogram
x5 = np.random.randn(1000)
plt.hist(x5,bins=10,color="r")
```

```python
d = np.linspace(5,20,5)
plt.bar(np.arange(len(d)),d)
plt.xlabel("length")
plt.ylabel("values of D")
plt.show()
```

```python
x6 = np.random.ranf(5)
labels = ["apple","bannana","orange","pineapple","kiwi"]
plt.pie(x6,labels=labels)
plt.title("Fruit share of market")
plt.show()
```
